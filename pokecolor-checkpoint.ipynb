# Calculating mean BGR value


```python
data_dict = {}
count =0
File_names = os.listdir(img_path)
print("Number of images: \n")
print(len(File_names))
```

```python
def mean_color(img):
    c=0
    colorSum = [0, 0, 0]
    for rc in img:
        for pixel in rc:
            if np.any(pixel>0):
                c+=1
                colorSum+=pixel
    return((colorSum/c/255))
```

```python
img = cv2.imread('archive/images/pokemon/abra.png')
meanColor=mean_color(img)           

print(meanColor)
```

```python

for file in File_names:
    img = cv2.imread(os.path.join(img_path, file)) 
    meanColor=mean_color(img)  
    data_dict[file.split(".")[0]] = np.array(meanColor)
print(data_dict)
```

# Split data


```python
from sklearn import svm, datasets
import sklearn.model_selection as model_selection
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.svm import SVC
```

```python
y = final_csv.Type1 #labels
X = final_csv.drop('Type1', axis=1) #features
X = X.drop('Name', axis=1) #features
```

```python
y.head()
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2)
print("\nX_train:\n")
print(X_train.head())
print(X_train.shape)

print("\nX_test:\n")
print(X_test.head())
print(X_test.shape)

print("\nY_train:\n")
print(y_train.head())
print(y_train.shape)

print("\nY_test:\n")
print(y_test.head())
print(y_test.shape)
```

# Build SVM with basic parameters

```python
rbf = svm.SVC(kernel='rbf').fit(X_train, y_train)
poly = svm.SVC(kernel='poly').fit(X_train, y_train)
```

```python
poly_pred = poly.predict(X_test)
rbf_pred = rbf.predict(X_test)
```

```python
poly_accuracy = accuracy_score(y_test, poly_pred)
poly_f1 = f1_score(y_test, poly_pred, average='weighted')
print('Accuracy (Polynomial Kernel): ', "%.2f" % (poly_accuracy*100))
print('F1 (Polynomial Kernel): ', "%.2f" % (poly_f1*100))
```

```python
rbf = svm.SVC(kernel='rbf', gamma = 10, C = 100 ).fit(X_train, y_train)
rbf_pred = rbf.predict(X_test)
```

```python
rbf_accuracy = accuracy_score(y_test, rbf_pred)
rbf_f1 = f1_score(y_test, rbf_pred, average='weighted')
print('Accuracy (RBF Kernel): ', "%.2f" % (rbf_accuracy*100))
print('F1 (RBF Kernel): ', "%.2f" % (rbf_f1*100))
```

```python
g_par = [10, 20, 30, 50]
for par in g_par:
    poly = svm.SVC(kernel='poly', degree=3, gamma=par).fit(X_train, y_train)
    poly_pred = poly.predict(X_test)
    poly_accuracy = accuracy_score(y_test, poly_pred)
    poly_f1 = f1_score(y_test, poly_pred, average='weighted')
    print(par)
    print('Accuracy (Polynomial Kernel): ', "%.2f" % (poly_accuracy*100))
    print('F1 (Polynomial Kernel): ', "%.2f" % (poly_f1*100))
        
```
